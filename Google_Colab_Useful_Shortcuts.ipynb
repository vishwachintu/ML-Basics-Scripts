# Google Colab useful shortcuts

* to move cell up ctrl+m K
* to move cell down ctrl+m J

* to create a new cell below ctrl+m b
* to create a new cell above ctrl+m a

* to delete a cell ctrl+m d
* to convert a text cell to code cell ctrl + m + y
* to convert a code cellt o text cell ctrl + m + m (double tap m)
* to replace within cell ctrl + shift + h
* to replace within entire notebook ctrl + h



* ctrl + alt + P code snippets
* ctr + Shift + p command palette

```
"inline shortcut list"
"Shift + Alt + up copy current line above"
"Shift + Alt + down copy current line below"
"Alt + up move current line up"
"Alt + down move current line down"
# "Ctrl + / current line comment"
```

```
"Alt + Enter to run current cell and create new cell below"
"ctrl + enter to run current cell"
"ctrl+ shift + enter to run selection"
```

```
# ctrl + m + i to interrupt m
# ctrl + m + l to toggle line numbers
# ctrl + m + o to toggle output
while True:
  print ('1')
```

```
# ctrl + shift + l to select all occurence of a word
# alt + click to insert multiple cursors for editing multiple places at the same time
print('hello buddies boss')
print('hello buddies boss to the old boss')
print('hello buddies boss in this new boss')
```
